```python
import numpy as np 
from scipy import stats
# rng = np.random.default_rng()



nSims = 100000

M = 106

SD = 15

p = []

n = 26

# x = np.random.normal(M, SD, n)

# x


# mu = 106
# t_stat, p_value = stats.ttest_1samp(rng, mu)
# t_stat
```

```python
for i in range(nSims):
    x = np.random.normal(M, SD, n)
    z = stats.ttest_1samp(x, 100)
    p.append(z.pvalue)

p
```

```python
sum_pvals = 0
for ele in p:
    if ele < 0.05:
        sum_pvals = sum_pvals + ele

sum_pvals
```

```python
import statsmodels.stats.power as smp

eff_size = (M-100)/SD

smp.ttest_power(eff_size, nobs=n, alpha=.05, alternative='two-sided')

```
